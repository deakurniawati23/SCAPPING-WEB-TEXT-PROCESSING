```python
import requests
from bs4 import BeautifulSoup

url = 'https://id.wikihow.com/Berpakaian-dengan-Tepat'

page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
teks = soup.find_all('div', class_='headline_info')
teks
```

```python
teks = soup.find_all('span', class_='mw-headline')
teks
```

```python
id = ['Memilih Pakaian yang Terlihat Bagus','Memiliki Koleksi Pakaian yang Serbaguna','Mendapatkan Saran Lebih Lanjut']
```

```python
teks = [t.get_text(strip=True) for t in soup.find_all('span', class_='mw-headline')]
for item in teks:
    if item in id:
        print(f'{item}')
```
